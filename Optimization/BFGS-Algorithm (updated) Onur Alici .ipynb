```python
import numpy as np # To define matrices and vectors
import math # To use mathematical functions
from sympy import * # To define the derivation of differentiable functions
```

```python
def f(x_1, x_2):
    return 100 * (x_2 - x_1 ** 2) ** 2 + (1 - x_1) ** 2

x_1 = Symbol('x_1') # neccessary to define x_1 and thus to build the derivation of f after x_1
x_2 = Symbol('x_2') # neccessary to define x_2 and thus to build the derivation of f after x_2
# \u0394:= Symbol of Delta

print('\u0394f(x_1, x_2) / \u0394x_1 =\n', f(x_1, x_2).diff(x_1))
print('')
print('\u0394f(x_1, x_2) / \u0394x_2 =\n', f(x_1, x_2).diff(x_2))

# And thus we can define the Gradient of f now:

def Gradient_f(x_1, x_2):
    return (-400*x_1*(-x_1**2 + x_2) + 2*x_1 -2, -200*x_1**2 + 200*x_2)

def Gradient_f_1(x_1, x_2):
    return -400*x_1*(-x_1**2 + x_2) + 2*x_1 -2

def Gradient_f_2(x_1, x_2):
    return -200*x_1**2 + 200*x_2
```

```python
print('Now to the BFGS-Algorithm:')
```

```python
# First Example with x_old = (1.2, 1.2) 
(x_1, x_2) = (1.2, 1.2) # starting point
e = 10 ** (-4) # convergance tolerance
H_old = np.eye(2) # inverse Hessian approximation, here the identity matrix
H_new = np.eye(2)
(y_1, y_2) = (1.2, 1.2)
c_1 = 0.1 # for the Wolfe conditions
c_2 = 0.9 # for the Wolfe conditions
print('All calculated values of (x_1, x_2):\n')

k = 0

while k >=0 and np.linalg.norm(Gradient_f(x_1, x_2)) > e:# minimum condition
    H_old = H_new.copy()
    (x_1, x_2) = (y_1, y_2)
    p = -1 * H_new.reshape(2,2) @ np.array([Gradient_f_1(x_1, x_2), Gradient_f_2(x_1, x_2)]) # the search direction
    a = 0.9 # alpha_0 (k = 0)
    while f(x_1 + a * p[0], x_2 + a * p[1]) > f(x_1, x_2) + c_1 * a * np.array([Gradient_f_1(x_1, x_2), Gradient_f_2(x_1, x_2)]) @ p and np.array([Gradient_f_1(x_1 + a * p[0], x_2 + a * p[1]), Gradient_f_2(x_1 + a * p[0], x_2 + a * p[1])]) @ p < c_2 * np.array([Gradient_f_1(x_1, x_2), Gradient_f_2(x_1, x_2)]) @ p:
        a = a * 0.5 # alpha_k
        if f(x_1 + a * p[0], x_2 + a * p[1]) <= f(x_1, x_2) + c_1 * a * np.array([Gradient_f_1(x_1, x_2), Gradient_f_2(x_1, x_2)]) @ p and np.array([Gradient_f_1(x_1 + a * p[0], x_2 + a * p[1]), Gradient_f_2(x_1 + a * p[0], x_2 + a * p[1])]) @ p >= c_2 * np.array([Gradient_f_1(x_1, x_2), Gradient_f_2(x_1, x_2)]) @ p:
            break
    (y_1, y_2) = (y_1, y_2) + a * p # x_(k+1)
    s = np.array([y_1, y_2]) - np.array([x_1, x_2]) # s_k
    y = np.array([Gradient_f_1(y_1, y_2), Gradient_f_2(y_1, y_2)]) - np.array([Gradient_f_1(x_1, x_2), Gradient_f_2(x_1, x_2)]) # y_k
    q = 1 / (y.T.reshape(1, 2) @ s.reshape(2, 1)) # Rho_k
    H_new = (np.eye(2) - q * s.reshape(2, 1) @ y.T.reshape(1, 2)) @ H_old @ (np.eye(2) - q * y.reshape(2, 1) @ s.T.reshape(1, 2)) + q * s.reshape(2, 1) @ s.T.reshape(1, 2) # H_(k+1)
    k = k + 1
    for i in range(k, k+1):
        print(i, (y_1, y_2))
             
print('')
print('H_new =\n', H_new)
print('')
print('Hessian matrix of f = \n', np.linalg.inv(H_new))
print('')
print('Minimum of f: (y_1, y_2) =\n', y_1, y_2)
print('')
print('Gradient_f(y_1, y_2) =\n', Gradient_f(y_1, y_2))
print('')
print('Minimum of f is:')
print('')
print('f(y_1, y_2) =\n', f(y_1, y_2))
```

```python
# Next Example with x_old = (-1.2, 1) 
(x_1, x_2) = (-1.2, 1) # starting point
e = 10 ** (-4) # convergance tolerance
H_old = np.eye(2) # inverse Hessian approximation, here the identity matrix
H_new = np.eye(2)
(y_1, y_2) = (-1.2, 1)
c_1 = 0.1 # for the Wolfe conditions
c_2 = 0.9 # for the Wolfe conditions
print('All calculated values of (x_1, x_2):\n')

k = 0

while k >=0 and np.linalg.norm(Gradient_f(x_1, x_2)) > e: # minimum condition
    H_old = H_new.copy()
    (x_1, x_2) = (y_1, y_2)
    p = -1 * H_new.reshape(2,2) @ np.array([Gradient_f_1(x_1, x_2), Gradient_f_2(x_1, x_2)]) # the search direction
    a = 0.9 # alpha_0 (k = 0)
    while f(x_1 + a * p[0], x_2 + a * p[1]) > f(x_1, x_2) + c_1 * a * np.array([Gradient_f_1(x_1, x_2), Gradient_f_2(x_1, x_2)]) @ p and np.array([Gradient_f_1(x_1 + a * p[0], x_2 + a * p[1]), Gradient_f_2(x_1 + a * p[0], x_2 + a * p[1])]) @ p < c_2 * np.array([Gradient_f_1(x_1, x_2), Gradient_f_2(x_1, x_2)]) @ p:
        a = a * 0.5 # alpha_k
        if f(x_1 + a * p[0], x_2 + a * p[1]) <= f(x_1, x_2) + c_1 * a * np.array([Gradient_f_1(x_1, x_2), Gradient_f_2(x_1, x_2)]) @ p and np.array([Gradient_f_1(x_1 + a * p[0], x_2 + a * p[1]), Gradient_f_2(x_1 + a * p[0], x_2 + a * p[1])]) @ p >= c_2 * np.array([Gradient_f_1(x_1, x_2), Gradient_f_2(x_1, x_2)]) @ p:
            break
    (y_1, y_2) = (y_1, y_2) + a * p # x_(k+1)
    s = np.array([y_1, y_2]) - np.array([x_1, x_2]) # s_k
    y = np.array([Gradient_f_1(y_1, y_2), Gradient_f_2(y_1, y_2)]) - np.array([Gradient_f_1(x_1, x_2), Gradient_f_2(x_1, x_2)]) # y_k
    q = 1 / (y.T.reshape(1, 2) @ s.reshape(2, 1)) # Rho_k
    H_new = (np.eye(2) - q * s.reshape(2, 1) @ y.T.reshape(1, 2)) @ H_old @ (np.eye(2) - q * y.reshape(2, 1) @ s.T.reshape(1, 2)) + q * s.reshape(2, 1) @ s.T.reshape(1, 2) # H_(k+1)
    k = k + 1
    for i in range(k, k+1):
        print(i, (y_1, y_2))
        
print('') 
print('H_new =\n', H_new)
print('')
print('Hessian matrix of f = \n', np.linalg.inv(H_new))
print('')
print('Minimum of f: (y_1, y_2) =\n', y_1, y_2)
print('')
print('Gradient_f(y_1, y_2) =\n', Gradient_f(y_1, y_2))
print('')
print('Minimum of f is:')
print('')
print('f(y_1, y_2) =\n', f(y_1, y_2))
```
