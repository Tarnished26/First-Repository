```python
print('Aufgabe 5')

import numpy as np

def f(x):
    val = 0
    if x != 0:
        val = (1-np.cos(x))/(x*x);
        return val
    
from sympy import *

e_A = Symbol ('e_A')
a = Symbol('a')
h = 10 ** (-5)
p = 11
y = 0.0001

print('1. Spalte:')
def g(x):
    return f(x) + e_A * (a + i) # delta_i = a + i, sodass a + i in (-1;1) liegt mit i = 0,...,11
for i in range(0, p+1):
    print(g(y + i * h)) # Delta^0 Phi^_i
    
print('')
print('')

print('2. Spalte:')
def h_1(i):
    return (f(y + (i+1) *h) + e_A * (a + i+1)) - (f(y + i * h) + e_A * (a+i)) # Delta^1 Phi^_i
for i in range(0, 11):
    print(h_1(i))
    
print('')
print('')

print('3. Spalte:')
def h_2(i):
    return h_1(i + 1) - h_1(i) # Delta^2 Phi^_i
for i in range(0, 10):
    print(h_2(i))
    
print('')
print('')

print('4. Spalte:')
def h_3(i):
    return h_2(i + 1) - h_2(i) # Delta^3 Phi^_i
for i in range(0, 9):
    print(h_3(i))
    
print('')
print('')

print('5. Spalte:')
def h_4(i):
    return h_3(i + 1) - h_3(i) # Delta^4 Phi^_i
for i in range(0, 8):
    print(h_4(i))
    
print('')
print('')

print('6. Spalte:')
def h_5(i):
    return h_4(i + 1) - h_4(i) # Delta^5 Phi^_i
for i in range(0, 7):
    print(h_5(i))
    
print('')
print('')

print('7. Spalte:')
def h_6(i):
    return h_5(i + 1) - h_5(i) # Delta^6 Phi^_i
for i in range(0, 6):
    print(h_6(i))
    
print('')
print('')

print('8. Spalte:')
def h_7(i):
    return h_6(i + 1) - h_6(i) # Delta^7 Phi^_i
for i in range(0, 5):
    print(h_7(i))
    
print('')
print('')

print('9. Spalte:')
def h_8(i):
    return h_7(i + 1) - h_7(i) # Delta^8 Phi^_i
for i in range(0, 4):
    print(h_8(i))
    
print('')
print('')

print('10. Spalte:')
def h_9(i):
    return h_8(i + 1) - h_8(i) # Delta^9 Phi^_i
for i in range(0, 3):
    print(h_9(i))
    
print('')
print('')

print('11. Spalte:')
def h_10(i):
    return h_9(i + 1) - h_9(i) # Delta^10 Phi^_i
for i in range(0, 2):
    print(h_10(i))
    
print('')
print('')

print('12. Spalte:')
def h_11(i):
    return h_10(i + 1) - h_10(i) # Delta^11 Phi^_i
for i in range(0, 1):
    print(h_11(i))
```

```python
# ohne e_A und delta_i

print('Aufgabe 5')

import numpy as np

def f(x):
    val = 0
    if x != 0:
        val = (1-np.cos(x))/(x*x);
        return val
    

h = 10 ** (-5)
p = 11
y = 0.0001

print('1. Spalte:')
def g(x):
    return f(x) 
for i in range(0, p+1):
    print(g(y + i * h)) # Delta^0 Phi^_i
    
print('')
print('')

print('2. Spalte:')
def h_1(i):
    return f(y + (i+1) *h) - f(y + i * h) # Delta^1 Phi^_i
for i in range(0, 11):
    print(h_1(i))
    
print('')
print('')

print('3. Spalte:')
def h_2(i):
    return h_1(i + 1) - h_1(i) # Delta^2 Phi^_i
for i in range(0, 10):
    print(h_2(i))
    
print('')
print('')

print('4. Spalte:')
def h_3(i):
    return h_2(i + 1) - h_2(i) # Delta^3 Phi^_i
for i in range(0, 9):
    print(h_3(i))
    
print('')
print('')

print('5. Spalte:')
def h_4(i):
    return h_3(i + 1) - h_3(i) # Delta^4 Phi^_i
for i in range(0, 8):
    print(h_4(i))
    
print('')
print('')

print('6. Spalte:')
def h_5(i):
    return h_4(i + 1) - h_4(i) # Delta^5 Phi^_i
for i in range(0, 7):
    print(h_5(i))
    
print('')
print('')

print('7. Spalte:')
def h_6(i):
    return h_5(i + 1) - h_5(i) # Delta^6 Phi^_i
for i in range(0, 6):
    print(h_6(i))
    
print('')
print('')

print('8. Spalte:')
def h_7(i):
    return h_6(i + 1) - h_6(i) # Delta^7 Phi^_i
for i in range(0, 5):
    print(h_7(i))
    
print('')
print('')

print('9. Spalte:')
def h_8(i):
    return h_7(i + 1) - h_7(i) # Delta^8 Phi^_i
for i in range(0, 4):
    print(h_8(i))
    
print('')
print('')

print('10. Spalte:')
def h_9(i):
    return h_8(i + 1) - h_8(i) # Delta^9 Phi^_i
for i in range(0, 3):
    print(h_9(i))
    
print('')
print('')

print('11. Spalte:')
def h_10(i):
    return h_9(i + 1) - h_9(i) # Delta^10 Phi^_i
for i in range(0, 2):
    print(h_10(i))
    
print('')
print('')

print('12. Spalte:')
def h_11(i):
    return h_10(i + 1) - h_10(i) # Delta^11 Phi^_i
for i in range(0, 1):
    print(h_11(i))
    
    
```

```python
print('Aufgabe 6')
print('')
print('k = 4 (5.Spalte) ist die erste Spalte mit alternierendem Vorzeichen.')

import math
# Binomialkoeffizient: math.comb(n, k), mit n >= k

print('Fall 1: k = 4 (5.Spalte)')
print('')

def s_1(i):
    return abs(h_4(i)) / (math.comb(8, 4)) ** 0.5

for i in range(0,8):
    print(s_1(i))
print('')
print('=> max(s_1(i)) = e_A = s_1(0)')

print('')
print('')
print('Fall 2: k = 5 (6.Spalte)')
print('')

def s_2(i):
    return abs(h_5(i)) / (math.comb(10, 5)) ** 0.5

for i in range(0,7):
    print(s_2(i))
print('')
print('=> max(s_2(i)) = e_A = s_2(0)')

print('')
print('')
print('Fall 3: k = 6 (7.Spalte)')
print('')

def s_3(i):
    return abs(h_6(i)) / (math.comb(12, 6)) ** 0.5

for i in range(0,6):
    print(s_3(i))
print('')
print('=> max(s_3(i)) = e_A = s_3(0)')
```
