```python
# Inverse Iteration # Am Beispiel Lambda_1 = 4
import numpy as np 

eps = 10 ** (-15) # oder auch 10 ** (-15)
A = np.array([[0, 11, -5], [-2, 17, -7], [-4, 26, -10]])
n, n = A.shape
L = [4, 2, 1] # Eigenwerte der Matrix A
x_old = np.ones(n) # Startvektor
x_new = np.ones(n) 
x_tilde = np.ones(n)
a = 3.945432 # Shift bzgl. Eigenwert L[0] = 4
C = a * np.eye(n)
A_tilde = A - C 
k = 1

while k <= 1000:
    k = k + 1
    x_old = x_new.copy()
    x_tilde = np.linalg.solve(A_tilde, x_old)
    v = max(abs(x_tilde)) # Maximumsnorm
    x_new = (1 / v) * x_tilde
    B = A_tilde @ x_new
    i = max(B).argmax()
    c = ( x_new[i] / B[i] ) ** (-1) + a # Eigenwertnäherung # Mu^(k)
    if abs(c-4) < eps:
        break



print(c)
print('')
print('A @ x_new=\n', A @ x_new)
print('')
print('L[0] * x_new=\n', L[0] * x_new)
print('')
print('c * x_new=\n', c * x_new)
print('')
print('c = L[0]. => c ist ein Eigenwert von A mit Eigenvektor [1.6 2.4 4. ].')
```
