```python
import numpy as np # Beispiel: n = 3 <=> h^(-1) = 4

n = 3
h = 1 / (n+1)
b = h ** 2 * np.ones(n) 
A = np.zeros((n,n))
n, n = A.shape
for i in range(0, n):
    for j in range(0, n):
        if j == i:
            A[i][j] = 2
        if j - i == 1 or i - j == 1:
            A[i][j] = -1
        if j - i > 1 or i - j > 1:
            A[i][j] = 0


# Zum SOR-Verfahren: Fall 1: w = 1 (Gauß-Seidel-Verfahren)

x_new = np.zeros(n)
x_old = np.zeros(n)
k = 0

while max(abs(A @ x_new - b)) > max(abs(x_new)) * 10 ** (-8) and k <= 1000:
    k = k + 1
    x_old = x_new.copy()
    for i in range(0, n):
        x_new[i] = b[i]
        for j in range(0, n):
            if j == i:
                continue
            if j < i:
                x_new[i] = x_new[i] - A[i][j] * x_new[j]
            if j > i:
                x_new[i] = x_new[i] - A[i][j] * x_old[j]
        x_new[i] = x_new[i] / A[i][i]

print(x_new)
```

```python
import numpy as np # Beispiel: n = 3 <=> h^(-1) = 4
from math import pi

n = 3
h = 1 / (n+1)
b = h ** 2 * np.ones(n)
A = np.zeros((n,n))
n, n = A.shape
for i in range(0, n):
    for j in range(0, n):
        if j == i:
            A[i][j] = 2
        if j - i == 1 or i - j == 1:
            A[i][j] = -1
        if j - i > 1 or i - j > 1:
            A[i][j] = 0


# Zum SOR-Verfahren: Fall 1: w = w_opt (SOR-Verfahren)
q = pi
p = 1 - (q ** 2 / 2) * h ** 2
w = 2 / (1 + (1 - p ** 2) ** 0.5)
x_new = np.zeros(n)
x_old = np.zeros(n)
k = 0

while max(abs(A @ x_new - b)) > max(abs(x_new)) * 10 ** (-8) and k <= 1000:
    k = k + 1
    x_old = x_new.copy()
    for i in range(0, n):
        x_new[i] = b[i]
        for j in range(0, n):
            if j == i:
                continue
            if j < i:
                x_new[i] = x_new[i] - A[i][j] * x_new[j]
            if j > i:
                x_new[i] = x_new[i] - A[i][j] * x_old[j]
        x_new[i] = w * x_new[i] / A[i][i] + (1-w) * x_old[i]


print(x_new)
```
