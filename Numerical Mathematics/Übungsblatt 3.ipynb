```python
# Erste Matrix der 13. Aufgabe

import numpy as np

e = 0.5
b = np.array([[1.0],[2.0]])
A = np.array([[e, 1], [1, 1]])

m, n = A.shape


print(A)

# Nun zum Verfahren: LGS(A|b) lösen, dabei Matrix A in eine obere Dreiecksmatrix bringen
for k in range(1, n):
    for i in range(k, n):
        factor = A[i][k-1] / A[k-1][k-1]
        for j in range(k-1, n):
            A[i][j] += - A[k-1][j] * factor
        b[i] += - b[k-1] * factor


# Rückwärtselimination: Bestimmung der x_i, i = 1,...,n
x = np.zeros(n)
x[n-1] = b[n-1] / A[n-1][n-1]

for i in range(n-2, -1, -1):
    x[i] = b[i]
    for j in range(i+1, n):
        x[i] = x[i] - A[i][j] * x[j]
    x[i] = x[i] / A[i][i]
    
print('')
print('A=\n', A)
print('b=\n', b)
print('')
print('Die Lösungen lauten:')
print('x =', x)
```
