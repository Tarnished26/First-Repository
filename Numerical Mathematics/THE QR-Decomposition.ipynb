```python
import numpy as np

A = np.array([[1, 2, 0.1, 3, 0.3], [-2, 4, 2, -9.8, 0], [100, 5, 3, 346, 0], 
              [0.5, 9, -0.5, 1.75, 0.3], [2, 6, 6, -0.2, 0], [6, 8, 8, 11.4, 0.3]])
Q = np.eye(6)
R = A

m, m = Q.shape
m, n = A.shape
m, n = R.shape

for k in range(0, m-1):
    x = np.zeros(m)
    for i in range(k, m):
        x[i] = R[i][k]
    g = np.linalg.norm(x)
    v = x
    v[k] = x[k] + g
    s = np.linalg.norm(v)
    if s!=0:
        w = v / s
        u = 2 * R.T.reshape(5,6) @ w.reshape(6,1)
        R = R - w.reshape(6,1) @ u.T.reshape(1,5)
        Q = Q - 2 * Q @ w.reshape(6,1) @ w.T.reshape(1,6)
        

print('Q=\n', Q)
print('')
print('')
print('R=\n', R)
print('')
print('R ist also eine obere Dreiecksmatrix.')
print('')
print('')
print('Q @ R =\n', Q.reshape(6,6) @ R.reshape(6,5))
print('')
print('Es gilt also: Q @ R = A.')
print('')
print('Q @ Q.T=\n', Q @ Q.T)
print('')
print('Q ist also orthogonal.')
```

```python
import numpy as np

b = np.array([[2.], [4.], [-2.], [10.]])
A = np.array([[1, -0.5, 0.5], [-1, 1, 1], [2, 4, 6], [-2, -4, -4]])
Q = np.eye(4)
R = A

m, m = Q.shape
m, n = A.shape
m, n = R.shape

for k in range(0, m-1):
    x = np.zeros(m)
    for i in range(k, m):
        x[i] = R[i][k]
    g = np.linalg.norm(x)
    v = x
    v[k] = x[k] + g
    s = np.linalg.norm(v)
    if s!=0:
        w = v / s
        u = 2 * R.T.reshape(3,4) @ w.reshape(4,1)
        R = R - w.reshape(4,1) @ u.T.reshape(1,3)
        Q = Q - 2 * Q @ w.reshape(4,1) @ w.T.reshape(1,4)
        

# Rückwärtselimination: Bestimmung der y_i, i = 1,2,3
y = np.zeros(n)
c = Q.T.reshape(4,4) @ b.reshape(4,1)

y[n-1] = c[n-1] / R[n-1][n-1]

for i in range(n-2, -1, -1):
    y[i] = c[i]
    for j in range(i+1, n):
        y[i] = y[i] - R[i][j] * y[j]
    y[i] = y[i] / R[i][i]      

print('Q=\n', Q)
print('')
print('')
print('R=\n', R)
print('')
print('R ist also eine obere Dreiecksmatrix.')
print('')
print('')
print('Q @ R =\n', Q.reshape(4,4) @ R.reshape(4,3))
print('')
print('Es gilt also: Q @ R = A.')
print('')
print('Q @ Q.T=\n', Q @ Q.T)
print('')
print('Q ist also orthogonal.')
print('')
print('')
print('Lösung des LGS Ay=b:')
print('1.Schritt: Bestimme c aus Qc=b:')
print('Qc=b <=> c=Q.T @ b=\n', Q.T.reshape(4,4) @ b.reshape(4,1))
print('')
print('Die Lösung ist y=\n', y)
print('')
print('R @ y =\n', R.reshape(4,3) @ y.reshape(3,1))
print(np.linalg.lstsq(A,b))
```
