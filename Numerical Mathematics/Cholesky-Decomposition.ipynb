```python
# Algorithmus: Choleskyzerlegung
# Aufgabenteil a)

import numpy as np

A = np.array([[25, 15, -5], [15, 18, 0], [-5, 0, 11]])
D = np.zeros(3)
n, n = A.shape
R = np.zeros((3,3))

for i in range(0, n):
    D[i] = A[i][i]
    for k in range(0, i):
        D[i] = D[i] - R[k][i] * R[k][i]
    if D[i] <= 0:
        print('A nicht positiv definit.')
        break
    else:
        R[i][i] = D[i] ** 0.5
        for j in range(i+1, n):
            R[i][j] = A[i][j]
            for k in range(0, i):
                R[i][j] = R[i][j] - R[k][i] * R[k][j]
            R[i][j] = R[i][j] / R[i][i]


print('R.T =\n', R.T)
print('')
print('R =\n', R)
print('')
print('R.T @ R =\n', R.T.reshape(3,3) @ R.reshape(3,3))
print('')
print('A =\n', A)
print('')
print('Es gilt also: A = R.T @ R')
```
