```python
import numpy as np
import matplotlib.pyplot as plt
from copy import deepcopy
```

```python
print('Beispiele von Brownschen Bewegungen (W_t)_{t in \mathbb{R}_+}')
```

```python
mean = 0                                             # Erwartungswert von W_x                                    

N = 100000                                           # generiert N Punkte
dt = 1/N                                             # Zeitintervall = [0,1]

data = [(0,0)]                                       # unsere 2-dimensionale Liste, bestehend aus Zeit x und Wert W_x
x = 0                                                # W_0 = 0 f.s.
for t in range(N):
    dx = np.random.normal(mean, np.sqrt(dt))  
    x = x + dx                                       # berechnet W_x schrittweise
    data.append((dt*t+dt, x+dx))
```

```python
data = np.array(data)                                # erzeugt aus data eine Matrix

print(data)

f = plt.figure()                                     # ändert das Format der Abbildung
f.set_figwidth(8)                                    # hier die Weite
f.set_figheight(8)                                   # hier die Höhe
plt.plot(data[:, 0], data[:, 1], linewidth=0.5)      # beschreibt die Dicke der Linie der BBe:
plt.scatter(data[0, 0], data[0, 1],marker="^",color='r',label="Origin") # zeigt den Start der BBe an
plt.xlabel('Zeit t')
plt.ylabel('Wert W_t')
#plt.legend()                                         # zeigt 'Origin' auf der Grafik an
plt.title("Brownian Motion")
plt.show()
```

```python
print('Jetzt hier mit Symmetrie:')

print('')

print('Symmetrie B = -W:')

print('')

data_old = deepcopy(data)

for i in range(N+1):
    data[i,1] = data[i,1] * (-1)

print(data)

f = plt.figure()                                     
f.set_figwidth(8)
f.set_figheight(8)
plt.plot(data_old[:, 0], data_old[:, 1], linewidth=0.5)   
plt.plot(data[:, 0], data[:, 1], linewidth=0.5)       
plt.scatter(data[0, 0], data[0, 1],marker="^",color='r',label="Origin") 
plt.xlabel('Zeit t')
plt.ylabel('Wert B_t')
plt.legend()                            
plt.title("Brownian Motion")
plt.show()
```

```python
print('BBe mit Drift \mu und Volatilität \sigma, wobei \mu = -0,5 und \sigma = 0.01:')

print('')

data = [(0,0)]
x = 0                                                
for t in range(N):
    dx = np.random.normal(mean + (-0.5)*dt, 0.01 * np.sqrt(dt))  
    x = x + dx                                       
    data.append((dt*t + dt, x+dx))

data = np.array(data)                               

print(data)

f = plt.figure()                                     
f.set_figwidth(10)
f.set_figheight(10)
plt.plot(data[:, 0], data[:, 1], linewidth=0.5)       
plt.scatter(data[0, 0], data[0, 1],marker="^",color='r',label="Origin") 
plt.xlabel('Zeit t')
plt.ylabel('Wert W_t')
plt.legend()                            
plt.title("Brownian Motion")
#plt.xlim(0,0.01)
#plt.ylim(-0.01,0.01)
plt.show()
#plt.xlim(0,1)
```
